## 속성 데이터

- Survival - 생존여부: 0 = No, 1 = Yes
- Pclass - 티켓 등급: 1 = 1st, 2 = 2nd, 3 = 3rd
- Sex - 성별: male, female
- Age - 나이
- SibSp - 동승한 형제, 배우자 수
- Parch - 부모와 자녀의 수
- Ticket - 티켓 번호
- Fare - 승선 요금
- Cabin - 캐빈(객실) 번호
- Embarked - 승선한 항구: C = Cherbourg, Q = Queenstown, S = Southampton

```python
import pandas as pd
```

```python
train = pd.read_csv("train.csv",index_col=["PassengerId"])
train.head()
```

```python
train.columns
```

```python
test = pd.read_csv("test.csv", index_col=["PassengerId"])
test.head()
```

```python
test.columns
```

```python
# 성별을 구분하는 컬럼 추가 

train.loc[train["Sex"] == "male", "Sex_encoder"] = 0
train.loc[train["Sex"] == "female", "Sex_encoder"] = 1

train.head()
```

```python
test.loc[test["Sex"] == "male", "Sex_encoder"] = 0
test.loc[test["Sex"] == "female", "Sex_encoder"] = 1
        
print(test.shape)
test.head()
```

```python
train.head()
```

```python
test.head()
```

### Encode Embarked

```python
train_embarked = pd.get_dummies(train["Embarked"], prefix="Embarked")

print(train_embarked.shape)
train_embarked.head()
```

```python
test_embarked = pd.get_dummies(test["Embarked"], prefix="Embarked")

print(test_embarked.shape)
test_embarked.head()
```

```python
train = pd.concat([train, train_embarked], axis=1)

print(train.shape)
train.head()
```

```python
test = pd.concat([test, test_embarked], axis= 1)

test.head()
```

## 요금이 비어 있는 항목을 찾아본다

```python
test[pd.isnull(test["Fare"])]
```

```python
# 비어 있는 요금 항목을, 요금의 평균치로 대체한다
mean_fare = train["Fare"].mean()
print("Fare(Mean) = ${0:.3f}".format(mean_fare))
```

```python
test.loc[pd.isnull(test["Fare"]), "Fare"] = mean_fare

test[pd.isnull(test["Fare"])]
```

### Train

```python
feature_names = ["Pclass", "Sex_encoder", "Embarked_C","Embarked_Q","Embarked_S","Fare","Parch","SibSp"]

X_train = train[feature_names]

print(X_train.shape)
X_train.head()
```

```python
label_name = ["Survived"]

y_train = train[label_name]
y_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=5)
```

```python
tree.fit(X_train, y_train)
```

```python
X_test = test[feature_names]

print(X_test.shape)
X_test.head()
```

```python
y_pred = tree.predict(X_test)

print(y_pred.shape)
y_pred
```

## Submit

```python
import numpy as np

Submission = pd.read_csv("gender_submission.csv",index_col="PassengerId")

Submission["Survived"] = y_pred.astype(np.int32)

print(Submission.shape)
Submission.head()
```

```python
Submission.to_csv("D_Tree1.csv")
```

### Score = 0.78468

## 주요 속성 찾기

```python
import matplotlib.pyplot as plt
```

```python
tree.feature_importances_
```

```python
feature_names
```

```python
list(zip(feature_names, tree.feature_importances_.round(4)))
```

```python
df = pd.DataFrame({'feature': feature_names,'importance': tree.feature_importances_ })

df=df.sort_values('importance', ascending=False) 
df
```

```python
x = df.feature
y = df.importance
ypos = np.arange(len(x))
plt.figure(figsize=(10,7)) 
plt.barh(x, y) 
plt.yticks(ypos, x) 
plt.xlabel('Importance') 
plt.ylabel('Variable') 
plt.xlim(0, 1) 
plt.ylim(-1, len(x)) 
plt.show()
```
